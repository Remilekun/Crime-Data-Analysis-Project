```python
#due to the largeness of the data, I am using 30000 samples for this project

dfNew4 = dfNew3[0:30000]
```

```python
dfNew4
```

```python
from dateutil.parser import parse
from datetime import datetime


tCol = dfNew4.Date

List = [(datetime.ctime(parse(x[0:-3])),x[-2:]) for x in tCol]
dayList = []
monthList = []
periodList = []
for row in List:
    day = row[0][0:4]
    month = row[0][4:7]
    if row[1]=='AM':
        period = 'Morning'
    elif row[1] =='PM' and int(row[0][11:13])<4:
        period = 'Afternoon'
    elif row[1] =='PM' and int(row[0][11:13])<6:
        period = 'Evening' 
    elif row[1] =='PM' and int(row[0][11:13])>5:
        period = 'Night'
    else:
        period = 'Unknown'
        
    dayList.append(day)
    monthList.append(month)
    periodList.append(period)
    
print(len(dayList), len(monthList), len(periodList))    

dfNew4['month'] = monthList
dfNew4['day'] = dayList
dfNew4['period']= periodList
dfNew4.head()
```

```python
dfNew4.day.unique()
```

```python
wkList = []
for day in list(dfNew4.day):
    if day in ['Sat ', 'Sun ']:
        wkList.append('Weekend')
    else:
        wkList.append('Weekday')
dfNew4['dayType'] = wkList  

dfNew4
```
